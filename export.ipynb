```python
CFG_MobileMamba_T2 = {
        'img_size': 192,
        'embed_dim': [144, 272, 368],
        'depth': [1, 2, 2],
        'global_ratio': [0.8, 0.7, 0.6],
        'local_ratio': [0.2, 0.2, 0.3],
        'kernels': [7, 5, 3],
        'drop_path': 0,
        'ssm_ratio': 2,
    }
```

```python
def MobileMamba_T2(num_classes=1000, pretrained=False, distillation=False, fuse=False, pretrained_cfg=None, model_cfg=CFG_MobileMamba_T2):
    model = MobileMamba(num_classes=num_classes, distillation=distillation, **model_cfg)
    return model
```

```python
model = MobileMamba_T2()
```

```python
dummy_input = torch.randn(1, 3, 224, 224)
```

```python
weight = torch.load('model_weights/mobilemamba_t2/ckpt_300.pth')
```

```python
model.load_state_dict(weight['net'])
```

```python
onnx_path = "onnx_model/mobilemamba_t2/mobilemam_t2.onnx"
torch.onnx.export(model, dummy_input, onnx_path, 
                      export_params=True, opset_version=12,
                      do_constant_folding=True, input_names=['input'],
                      output_names=['output'], dynamic_axes={'input': {0: 'batch_size'},
                                                              'output': {0: 'batch_size'}})
pt_path = "onnx_model/mobilemamba_t2/mobilemamba.pt"
torch.save(model, pt_path)
```
